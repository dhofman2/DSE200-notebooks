Now write three functions for matrix multiply $C=AB$ in each of the following styles:

1. By using nested for loops to impliment the naive algorithm ($C_{ij}=\sum_{k=0}^{m-1}A_{ik}B_{kj}$)
2. Using numpy's built in martrix multiplication  
3. Using Cython

The three methods should have the same answer

```python


def python_prod(A, B):

    A_n, A_m = A.shape
    B_n, B_m = B.shape
    assert A_m == B_n, "Non-conformable shapes."
    C = np.zeros((A_n, B_m))
    for i in xrange(A_n):
        for j in xrange(B_m):
            for k in xrange(A_m):
                C[i, j] += A[i, k] * B[k, j]
    return C

def numpy_prod(A, B):
    return A.dot(B)

print 'Nested Loops Solution'
print python_prod(a,b)

print 'Numpy Solution'
print numpy_prod(a,b)
```

```python
import numpy as np

def random_prod(dim_a, dim_b, dim_c):
    print "dimensions:", dim_a, dim_b, dim_c
    aa=np.matrix(np.random.random_integers(0,dim_a,(dim_a,dim_b)))
    bb=np.matrix(np.random.random_integers(0, dim_b, (dim_b,dim_c)))
    s=np.size(numpy_prod(aa,bb)) #get size of output matrix for x-axis of plot
    
    #print 'numpy'
    #print numpy_prod(aa,bb)[1,:5]
    t1 = time.time()
    numpy_prod(aa,bb)
    t2 = time.time()
    
    #print 'python'
    #print python_prod(aa,bb)[1,:5]
    t3 = time.time()
    python_prod(aa,bb)
    t4 = time.time()
    
    #print 'cython'
    #print cython_prod(aa,bb)[1,:5]
    t5 = time.time()
    cython_prod(aa, bb)
    t6 = time.time()

    return ([s, (t2-t1), (t4-t3), (t6-t5)])

ntimes=np.ndarray((8,2),float)
ptimes=np.ndarray((8,2),float)
ctimes=np.ndarray((8,2),float)
counter=0
for adim in [10,100]:
    for bdim in [10,100]:
        for cdim in [10,100]:
            tout=random_prod(adim,bdim,cdim) 
            ntimes[counter,0]=tout[0]
            ntimes[counter,1]=tout[1]
            ptimes[counter,0]=tout[0]
            ptimes[counter,1]=tout[2]
            ctimes[counter,0]=tout[0]
            ctimes[counter,1]=tout[3]
            counter+=1

numpoints = len(ntimes)
print numpoints
```

```python
import pickle
import gzip
import json
import pandas as pd
import sqlite3

#create database and get the pickle data
con = sqlite3.connect("datatype.sqlite")
datatypes=pickle.load(gzip.open('datatypes.pkl','rb'))

#create a script to execute to create the table and load the pickle data
cursor = con.cursor()
cursor.execute('''DROP TABLE IF EXISTS datatypes''')
con.commit()

cursor.execute("CREATE TABLE datatypes ('PythonType' TEXT KEY NOT NULL,'JSONType' TEXT)")
con.commit()

for key, value in datatypes.items():
    statement = "INSERT INTO datatypes VALUES('%12s', '%12s')"%(value,key)
    cursor.execute(statement)
                  
con.commit()

# read the values back to confirm
#df = pd.read_sql("SELECT * from datatypes LIMIT 3", con)
#df

#create missing datatypes table
cursor = con.cursor()
cursor.execute('''DROP TABLE IF EXISTS missing_datatypes''')
con.commit()

cursor.execute("CREATE TABLE missing_datatypes ('PythonType' TEXT KEY NOT NULL,'JSONType' TEXT)")
con.commit()

missing_datatypes = {'byte':'', 'tuple':''}

for key, value in missing_datatypes.items():
    statement = "INSERT INTO missing_datatypes VALUES('%12s', '%12s')"%(key,value)
    cursor.execute(statement)
                  
con.commit()

# read the values back to confirm
#df = pd.read_sql("SELECT * from missing_datatypes LIMIT 3", con)
#df

#merge the tables
script = "INSERT INTO datatypes \
SELECT * from missing_datatypes"
cursor.execute(script)
con.commit()

# read the sorted values back
df= pd.read_sql("SELECT * FROM datatypes ORDER BY trim(Lower(PythonType)) ASC", con)
df
```
